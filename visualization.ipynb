```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='claim_status', data=df)
plt.title('Fraud vs Non-Fraud Claims')
plt.xlabel('Claim Status (0 = Genuine, 1 = Fraud)')
plt.ylabel('Number of Claims')
plt.show()
```

```python
# Distribution of Customer Age by Claim Status

sns.histplot(data=df, x='customer_age', hue='claim_status', kde=True, element='step')
plt.title('Customer Age Distribution by Claim Status')
plt.xlabel('Customer Age')
plt.ylabel('Count')
plt.show()
```

```python
# Boxplot of Subscription Length by Claim Status

sns.boxplot(x='claim_status', y='subscription_length', data=df)
plt.title('Subscription Length vs Claim Status')
plt.xlabel('Claim Status')
plt.ylabel('Subscription Length (Months)')
plt.show()

```

```python
# Fuel Type vs Fraud
if 'fuel_type_Petrol' in df.columns:
    df['fuel_type'] = df[['fuel_type_Petrol', 'fuel_type_Diesel']].idxmax(axis=1).str.replace('fuel_type_', '')
    sns.countplot(x='fuel_type', hue='claim_status', data=df)
    plt.title('Fuel Type vs Claim Status')
    plt.show()
```
