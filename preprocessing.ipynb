```python

import pandas as pd

# Load the data
df = pd.read_csv("../data/insurance claims data.csv")

# Converting binary yes/no columns to 1/0
binary_cols = [col for col in df.columns if col.startswith("is_")]
for col in binary_cols:
    df[col] = df[col].map({'Yes': 1, 'No': 0})

# Droping the irrelevant columns (IDs or non-useful)
drop_cols = ['policy_id', 'model', 'max_torque', 'max_power']
df = df.drop(columns=drop_cols, errors='ignore')

# One-hot encode the categorical columns
df = pd.get_dummies(df, drop_first=True)

# Filling the missing values with column median
df = df.fillna(df.median(numeric_only=True))


# Save cleaned data to file
df.to_csv("../data/insurance claims data.csv", index=False)
print("Preprocessing is complete and saved ")

# Show preview
print(df.head())
```
